```python
import pandas as pd
from IPython.display import HTML

''' ABILITIES '''

pd.set_option("display.max_colwidth", 250)
df = pd.read_csv('shared\\abilities.csv', sep='\t')
tags = ['Огнестрельное оружие', 'Винтовки']
clear_match = False
if clear_match == True:
    df = df[df['Теги'].apply(lambda x: all(tag in x for tag in tags))]
else:
    df = df[df['Теги'].apply(lambda x: any(tag in x for tag in tags))]
display(df.sort_values(by=['Уровень'], ascending=True).style.hide_index())
```
